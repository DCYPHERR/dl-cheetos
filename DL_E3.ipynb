# **QUESTION 1**

**Create a single layer neural network to classify the self generated random data**

**Importing Libraries**

````python
import torch
import torch.nn as nn
````

**Data preparation**

````python
n_input, n_hidden, n_out, batch_size, learning_rate = 10, 1, 1, 100, 0.95 #change learning rate here 
````

**Making dummy input data and output data**

````python
data_x = torch.randn(batch_size, n_input)
data_y = (torch.rand(size=(batch_size, 1)) < 0.5).float()
````

**initialized the input data with 100 data samples with 10 features each and respectively initialized the output data with 100 data points.**

````python
print(data_x.size())
print(data_y.size())
````

**Neural Network Model**

````python
model = nn.Sequential(nn.Linear(n_input, n_hidden),
                      nn.ReLU(),
                      nn.Linear(n_hidden, n_out),
                      nn.Sigmoid())
print(model)
````



```
# add another hidden layer
import torch
import torch.nn as nn

n_input, n_hidden1, n_hidden2, n_out, batch_size, learning_rate = 10, 5, 3, 1, 100, 0.95

data_x = torch.randn(batch_size, n_input)
data_y = (torch.rand(size=(batch_size, 1)) < 0.5).float()
print(data_x.size())
print(data_y.size())

model = nn.Sequential(
    nn.Linear(n_input, n_hidden1),
    nn.ReLU(),
    nn.Linear(n_hidden1, n_hidden2),
    nn.ReLU(),
    nn.Linear(n_hidden2, n_out),
    nn.Sigmoid()
)

print(model)

```



**Defining the loss function and the optimizer**

````python
loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
````

**Training Stage**

````python
losses = []
for epoch in range(5000):
    pred_y = model(data_x)
    loss = loss_function(pred_y, data_y)
    losses.append(loss.item())

    model.zero_grad()
    loss.backward()

    optimizer.step()
````

````python
print(losses)
````

**Output**

````python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title("Learning rate %f"%(learning_rate))
plt.show()
````

**Study the single layer neural network for different Learning rates**

A single layer neural network is a simple feedforward network consisting of a single layer of neurons connected to the inputs and outputs. The learning rate determines the size of the step that the model takes towards minimizing the loss function during training.

Different learning rates can impact the performance of the model in different ways. If the learning rate is too high, the model may overshoot the optimal solution and fail to converge, resulting in oscillations or instability in the training process. If the learning rate is too low, the model may take a very long time to converge or get stuck in a suboptimal solution.

On the other hand, an appropriate learning rate can help the model converge to the optimal solution more quickly and efficiently. Experimenting with different learning rates is an important part of the hyperparameter tuning process in deep learning, as finding the right learning rate can greatly improve the performance of the model.
