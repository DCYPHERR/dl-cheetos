# **Question 3**

**The below implementation has a total of 2 hidden layers**

````python
import torch
import torch.nn as nn

n_input, n_hidden_1, n_hidden_2, n_out, batch_size, learning_rate = 10, 5, 2, 1, 100, 0.95
data_x = torch.randn(batch_size, n_input)
data_y = (torch.rand(size=(batch_size, 1)) < 0.5).float()
print(data_x.size())
print(data_y.size())

model = nn.Sequential(nn.Linear(n_input, n_hidden_1),
                      nn.ReLU(),
                      nn.Linear(n_hidden_1, n_hidden_2),
                      nn.ReLU(),
                      nn.Linear(n_hidden_2, n_out),
                      nn.Sigmoid())
print(model)

loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
losses = []
for epoch in range(5000):
    pred_y = model(data_x)
    loss = loss_function(pred_y, data_y)
    losses.append(loss.item())

    model.zero_grad()
    loss.backward()

    optimizer.step()
print(losses)

import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title("Learning rate %f"%(learning_rate))
plt.show()
````
